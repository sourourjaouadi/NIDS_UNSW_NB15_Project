```python
from sklearn.preprocessing import LabelEncoder

for col in ['proto', 'service', 'state']:
    le = LabelEncoder()
    X_train[col] = le.fit_transform(X_train[col])
    X_val[col]   = le.transform(X_val[col])
    X_test[col]  = le.transform(X_test[col])
```

```python
categorical_cols = ['srcip','sport','dstip','dsport','ct_ftp_cmd']

for col in categorical_cols:
    # Convert everything to string first
    X_train[col] = X_train[col].astype(str)

    # Then apply label encoding
    from sklearn.preprocessing import LabelEncoder
    le = LabelEncoder()
    X_train[col] = le.fit_transform(X_train[col])
```

```python
# Restore y_train, y_val, y_test to their original numeric Series from the splits
# This undoes the incorrect LabelEncoding of the target variable in cell jEIlc2haqaK8
y_bin_train = y_bin.loc[X_train.index]
y_bin_val = y_bin.loc[X_val.index]
y_bin_test = y_bin.loc[X_test.index]

# Columns to be dropped to avoid overfitting and data leakage
cols_to_drop = ['srcip', 'dstip', 'stime', 'ltime']

for df_split in [X_train, X_val, X_test]:
    for col_to_drop in cols_to_drop:
        if col_to_drop in df_split.columns:
            df_split.drop(col_to_drop, axis=1, inplace=True)

# Define columns that might be object or mixed and need to be strictly numeric
cols_to_convert_numeric = ['sport', 'dsport', 'ct_ftp_cmd']

# Ensure 'sport', 'dsport', 'ct_ftp_cmd' are numeric, handling NaNs
for col in cols_to_convert_numeric:
    # Convert to numeric, coercing errors will turn invalid parsing into NaN
    X_train[col] = pd.to_numeric(X_train[col], errors='coerce')
    X_val[col] = pd.to_numeric(X_val[col], errors='coerce')
    X_test[col] = pd.to_numeric(X_test[col], errors='coerce')

    # Fill NaN values. Using 0 as a placeholder for missing numeric values.
    X_train[col].fillna(0, inplace=True);
    X_val[col].fillna(0, inplace=True);
    X_test[col].fillna(0, inplace=True);

    # Ensure integer type after fillna for consistency
    X_train[col] = X_train[col].astype(int)
    X_val[col] = X_val[col].astype(int)
    X_test[col] = X_test[col].astype(int)
```
